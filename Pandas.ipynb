```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_vendas_clientes.json')
dados.head()
```

```python
#normalizando os dados
dados = pd.json_normalize(dados['dados_vendas'])
dados
```

```python
#visualizando o nome das colunas
colunas = list(dados.columns)
colunas
```

```python
# Usando o .explode() para desagrupar dados de uma lista
dados = dados.explode(colunas[1:])
dados
```

```python
# Resentando os 鱈ndices ap坦s realizar o desagrupamento dos dados
dados.reset_index(inplace=True,drop=True)
dados.head()
```

```python
# Verificando os tipos de dados
dados.info()
```

```python
import  numpy as np
```

```python
#Transformando a coluna numerica para tipo numerico removendo o cifr達o
dados['Valor da compra'] = dados['Valor da compra'].apply(lambda x: x.replace('R','').replace('$','').replace(',','').strip())
```

```python
dados['Valor da compra'] = dados['Valor da compra'].astype(np.float64)
```

```python
dados.info()
```

```python
dados['Cliente'] = dados['Cliente'].str.lower()
```

```python
dados.head()
```

```python
#Removendo caracteres especiais dos nomes
dados['Cliente'] = dados['Cliente'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
#Removendo os hifens que n達o estejam conectando palavras compostas, ou seja, os hifens soltos que n達o tenham nenhum caractere antes ou depois deles
dados['Cliente'] = dados['Cliente'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
```
